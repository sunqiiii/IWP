## Data Cleaning
You need to clean the rows and columns of data using Pandas.  Insert the code (and descriptions of the code in markdown) in this subsection.

```python
# insert code here
df_cars
```

```python
# Remove duplicate rows
# the code below is from chatgpt
df_cars.drop_duplicates(inplace=True)
df_cars = df_cars.drop_duplicates(subset='VIN#')
```

```python
# Remove rows with missing values
# the code below is from chatgpt
df_cars.dropna(inplace=True)
```

```python
# the code below is from chatgpt
# Define a function to remove non-numeric characters
def extract_numeric(value):
    if isinstance(value, str):
        numeric_part = ''.join(re.findall(r'\d+', value))
        if numeric_part:
            return int(numeric_part)
    return value

# Apply the function to the 'Mileage' column
df_cars['Mileage'] = df_cars['Mileage'].apply(extract_numeric)
df_cars['Price'] = df_cars['Price'].apply(extract_numeric)
df_cars['Doors'] = df_cars['Doors'].apply(extract_numeric)
```

```python
# the code below is from chatgpt
# adjust the year value to integer
df_cars['Year'] = df_cars['Year'].astype(int)
```

```python
# the code below is from chatgpt
# exclude the data beyond the year range
df_cars = df_cars[(df_cars['Year'] >= 2008) & (df_cars['Year'] <= 2022)]
```

## Export Cleaned Data
